````python
# Cell 8: Process Datasets 1,2,3,5 (Skip Dataset 4 for now)
print("üìö PROCESSING DATASETS 1,2,3,5 INTO UNIFIED CHUNKS...")
print("="*50)

all_chunks = []
all_metadata = []

def create_legal_chunks(texts, titles, source, chunk_size=500, overlap=50):
    chunks = []
    metadata = []

    for i, (text, title) in enumerate(zip(texts, titles)):
        if not text: continue

        sentences = text.split('. ')
        current_chunk = ""

        for sentence in sentences:
            if len(current_chunk) + len(sentence) > chunk_size and current_chunk:
                chunks.append(current_chunk.strip())
                metadata.append({'title': title, 'source': source})
                overlap_sentences = current_chunk.split('. ')[-2:]
                current_chunk = '. '.join(overlap_sentences) + '. ' + sentence + '. '
            else:
                current_chunk += sentence + '. '

        if current_chunk.strip():
            chunks.append(current_chunk.strip())
            metadata.append({'title': title, 'source': source})

    return chunks, metadata

# 1. Process ILC Dataset
print("üìñ Processing Dataset 1 - ILC Cases...")
ilc_cases = dataset1['train']['Case'][:] + dataset1['test']['Case'][:]
ilc_titles = dataset1['train']['Title'][:] + dataset1['test']['Title'][:]
ilc_chunks, ilc_meta = create_legal_chunks(ilc_cases, ilc_titles, "ILC Dataset")
all_chunks.extend(ilc_chunks)
all_metadata.extend(ilc_meta)
print(f"   ‚úÖ {len(ilc_chunks):,} chunks")

# 2. Process Indian Legal QA
print("üìñ Processing Dataset 3 - Indian Legal QA...")
dataset3_path = '/content/dataset3/'
qa_chunks = []
qa_meta = []

for file in ['constitution_qa.json', 'crpc_qa.json', 'ipc_qa.json']:
    file_path = os.path.join(dataset3_path, file)
    if os.path.exists(file_path):
        with open(file_path, 'r') as f:
            data = json.load(f)

        law_type = file.replace('_qa.json', '').upper()
        for i, item in enumerate(data):
            if 'question' in item and 'answer' in item:
                qa_text = f"Question: {item['question']} Answer: {item['answer']}"
                qa_chunks.append(qa_text)
                qa_meta.append({'title': f"{law_type} Q&A", 'source': "Indian Legal QA"})

all_chunks.extend(qa_chunks)
all_metadata.extend(qa_meta)
print(f"   ‚úÖ {len(qa_chunks):,} chunks")

# 3. Process IndicLegalQA
print("üìñ Processing Dataset 2 - IndicLegalQA...")
with open('/content/IndicLegalQA.json', 'r') as f:
    indic_data = json.load(f)

indic_chunks = []
indic_meta = []

for i, item in enumerate(indic_data):
    if 'question' in item and 'answer' in item:
        qa_text = f"Case: {item.get('case_name', 'Unknown')}. Question: {item['question']} Answer: {item['answer']}"
        indic_chunks.append(qa_text)
        indic_meta.append({'title': f"Indic: {item.get('case_name', 'Unknown')}", 'source': "IndicLegalQA"})

all_chunks.extend(indic_chunks)
all_metadata.extend(indic_meta)
print(f"   ‚úÖ {len(indic_chunks):,} chunks")

# 4. Process Legal Case Summarization
print("üìñ Processing Dataset 5 - Legal Case Summarization...")
dataset5_path = '/content/dataset5/dataset'
summary_chunks = []
summary_meta = []

in_abs_path = os.path.join(dataset5_path, 'IN-Abs')
if os.path.exists(in_abs_path):
    train_judgement_path = os.path.join(in_abs_path, 'train-data', 'judgement')

    if os.path.exists(train_judgement_path):
        files = os.listdir(train_judgement_path)[:2000]  # Sample 2000 files for speed
        for i, filename in enumerate(files):
            if filename.endswith('.txt'):
                judgement_file = os.path.join(train_judgement_path, filename)

                try:
                    with open(judgement_file, 'r', encoding='utf-8') as f:
                        judgement_text = f.read()

                    summary_chunks.append(judgement_text)
                    summary_meta.append({'title': f"Case: {filename}", 'source': "Case Summarization"})
                except:
                    pass

all_chunks.extend(summary_chunks)
all_metadata.extend(summary_meta)
print(f"   ‚úÖ {len(summary_chunks):,} chunks")

print(f"\nüéâ DATASETS 1,2,3,5 PROCESSED: {len(all_chunks):,} chunks")
````

````python
import json
# Cell 10: BUILD RAG PIPELINE WITH GEMINI
print("üöÄ BUILDING RAG PIPELINE WITH GEMINI...")
print("="*50)

# Load the saved FAISS index and metadata
print("üì• Loading FAISS index and metadata...")
index = faiss.read_index('/content/main_legal_index.faiss')

with open('/content/main_chunk_metadata.json', 'r') as f:
    all_metadata = json.load(f)

with open('/content/main_chunks.json', 'r') as f:
    all_chunks = json.load(f)

print(f"‚úÖ Loaded: {index.ntotal} vectors, {len(all_metadata)} metadata, {len(all_chunks)} chunks")

class LegalRAGSystem:
    def __init__(self, index, chunks, metadata, embed_model, llm_model, k=5):
        self.index = index
        self.chunks = chunks
        self.metadata = metadata
        self.embed_model = embed_model
        self.llm_model = llm_model
        self.k = k

    def retrieve(self, query: str) -> List[Dict]:
        """Retrieve most relevant legal chunks for a query"""
        query_embedding = self.embed_model.encode([query])
        distances, indices = self.index.search(query_embedding.astype('float32'), self.k)

        results = []
        for i, (dist, idx) in enumerate(zip(distances[0], indices[0])):
            if idx < len(self.chunks):  # Safety check
                results.append({
                    'chunk': self.chunks[idx],
                    'metadata': self.metadata[idx],
                    'score': float(dist),
                    'rank': i + 1
                })
        return results

    def format_context(self, retrieved_results: List[Dict]) -> str:
        """Format retrieved context for the LLM"""
        context_parts = []

        for i, result in enumerate(retrieved_results):
            source = result['metadata']['source']
            title = result['metadata']['title'][:100]  # Truncate long titles
            chunk_text = result['chunk']

            context_parts.append(f"[SOURCE {i+1}: {source} - {title}]\n{chunk_text}\n")

        return "\n".join(context_parts)

    def generate_answer(self, query: str, retrieved_results: List[Dict]) -> Dict:
        """Generate answer using Gemini with retrieved context"""
        context = self.format_context(retrieved_results)

        prompt = f"""You are an Indian Legal AI Assistant. Your role is to provide accurate, clear, and helpful answers about Indian law based ONLY on the provided legal context.

LEGAL CONTEXT:
{context}

USER QUESTION: {query}

INSTRUCTIONS:
1. Answer the question using ONLY the information from the legal context above
2. If the context doesn't contain relevant information, say "I cannot find specific information about this in the provided legal sources."
3. Provide clear, plain-language explanations suitable for non-lawyers
4. ALWAYS cite your sources by referring to the SOURCE numbers in brackets
5. Be precise about legal provisions, sections, and case laws when mentioned in context
6. If discussing punishments or legal consequences, specify the relevant sections

ANSWER:"""

        try:
            response = self.llm_model.generate_content(prompt)
            answer = response.text

            return {
                'answer': answer,
                'sources': retrieved_results,
                'context_used': len(retrieved_results)
            }

        except Exception as e:
            return {
                'answer': f"Error generating answer: {str(e)}",
                'sources': retrieved_results,
                'context_used': len(retrieved_results)
            }

    def query(self, question: str) -> Dict:
        """Main query function - retrieve and generate"""
        print(f"üîç Searching for: '{question}'")

        # Retrieve relevant chunks
        retrieved = self.retrieve(question)
        print(f"üìö Retrieved {len(retrieved)} relevant legal documents")

        # Generate answer
        result = self.generate_answer(question, retrieved)

        return result

# Initialize the RAG system
legal_rag = LegalRAGSystem(
    index=index,
    chunks=all_chunks,
    metadata=all_metadata,
    embed_model=embed_model,
    llm_model=gemini_model,
    k=5  # Top 5 most relevant chunks
)

print("‚úÖ Legal RAG System Initialized!")
print(f"üí° Ready to answer legal questions using {len(all_chunks):,} legal chunks")

# Test the system
print("\nüß™ TESTING THE RAG SYSTEM...")
print("="*50)

test_questions = [
    "What is the punishment for theft under Indian law?",
    "What are the fundamental rights in Indian Constitution?",
    "What is the procedure for filing a criminal case?",
    "What is the definition of murder in IPC?"
]

for i, question in enumerate(test_questions, 1):
    print(f"\nüìù Test {i}: {question}")
    result = legal_rag.query(question)

    print(f"ü§ñ ANSWER: {result['answer'][:300]}...")
    print(f"üìö Sources used: {result['context_used']}")

    # Show top source
    if result['sources']:
        top_source = result['sources'][0]
        print(f"üìñ Top source: {top_source['metadata']['source']} - {top_source['metadata']['title']}")

    print("-" * 80)
````

````python
# Cell 12: ADVANCED FEATURES - CITATION AND CONFIDENCE
print("üî¨ ADVANCED RAG FEATURES")
print("="*50)

class AdvancedLegalRAG(LegalRAGSystem):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def calculate_confidence(self, retrieved_results: List[Dict]) -> float:
        """Calculate confidence score based on retrieval quality"""
        if not retrieved_results:
            return 0.0

        # Convert distances to similarities (higher is better)
        similarities = [1 / (1 + result['score']) for result in retrieved_results]
        avg_similarity = np.mean(similarities)

        # Normalize to 0-1 scale
        confidence = min(avg_similarity * 2, 1.0)  # Adjust scaling as needed
        return round(confidence, 2)

    def extract_citations(self, answer: str, retrieved_results: List[Dict]) -> List[Dict]:
        """Extract and format citations from answer"""
        citations = []

        # Simple source matching - in production, use more sophisticated NLP
        for i, result in enumerate(retrieved_results):
            source_key = f"[SOURCE {i+1}]"
            if source_key in answer:
                citations.append({
                    'source_number': i + 1,
                    'document': result['metadata']['title'],
                    'source_type': result['metadata']['source'],
                    'content_preview': result['chunk'][:200] + "..."
                })

        return citations

    def query_with_analysis(self, question: str) -> Dict:
        """Enhanced query with confidence and citation analysis"""
        # Retrieve
        retrieved = self.retrieve(question)

        # Calculate confidence
        confidence = self.calculate_confidence(retrieved)

        # Generate answer
        result = self.generate_answer(question, retrieved)

        # Extract citations
        citations = self.extract_citations(result['answer'], retrieved)

        # Enhanced result
        enhanced_result = {
            **result,
            'confidence': confidence,
            'citations': citations,
            'retrieval_metrics': {
                'total_chunks_retrieved': len(retrieved),
                'average_similarity_score': np.mean([1/(1+r['score']) for r in retrieved]) if retrieved else 0
            }
        }

        return enhanced_result

# Initialize advanced RAG
advanced_rag = AdvancedLegalRAG(
    index=index,
    chunks=all_chunks,
    metadata=all_metadata,
    embed_model=embed_model,
    llm_model=gemini_model
)

print("‚úÖ Advanced Legal RAG System Ready!")

# Test advanced features
print("\nüß™ TESTING ADVANCED FEATURES...")
test_question = "What is the punishment for robbery under Indian Penal Code?"
result = advanced_rag.query_with_analysis(test_question)

print(f"üìù Question: {test_question}")
print(f"üéØ Answer: {result['answer'][:500]}...")
print(f"üìä Confidence: {result['confidence'] * 100}%")
print(f"üîó Citations found: {len(result['citations'])}")
print(f"üìö Retrieval metrics: {result['retrieval_metrics']}")

print("\nüìã Detailed Citations:")
for citation in result['citations']:
    print(f"  ‚Ä¢ Source {citation['source_number']}: {citation['document']}")
    print(f"    Type: {citation['source_type']}")
````

````python
import json
import numpy as np
import time
import collections
import re
from tqdm import tqdm

print("üöÄ STARTING OPTIMIZED EVALUATION (Combined Judging)...")
print("="*50)

# ==========================================
# 1. Retry Logic (Crucial for Rate Limits)
# ==========================================
def run_with_retry(func, *args, max_retries=5, initial_delay=2):
    delay = initial_delay
    for attempt in range(max_retries):
        try:
            return func(*args)
        except Exception as e:
            error_str = str(e)
            if "429" in error_str or "Resource exhausted" in error_str:
                print(f"‚è≥ Rate limit. Waiting {delay}s...")
                time.sleep(delay)
                delay *= 2
            else:
                print(f"‚ùå Error: {e}")
                return None
    return None

# ==========================================
# 2. Unified Judge Function (The Optimization)
# ==========================================
def combined_judge_call(question, answer, context):
    """
    Asks Gemini to grade Faithfulness AND Relevancy in ONE go.
    Returns a dictionary: {'faithfulness': 0.0, 'relevancy': 0.0}
    """
    prompt = f"""
    You are an expert legal AI evaluator. Grade the following AI Answer based on the User Question and Context.

    USER QUESTION: {question}

    RETRIEVED CONTEXT:
    {context}

    AI ANSWER:
    {answer}

    --- YOUR TASK ---
    Provide two scores (0.0 to 1.0):
    1. "faithfulness": Is the answer derived *only* from the Context? (1.0 = Yes, 0.0 = Hallucination)
    2. "relevancy": Does the answer address the User Question? (1.0 = Yes, 0.0 = Irrelevant)

    Return ONLY a JSON object. Do not write explanations.
    Format: {{"faithfulness": 0.9, "relevancy": 0.8}}
    """
    response = gemini_model.generate_content(prompt)
    text = response.text.strip()

    # Clean up markdown if Gemini adds it (e.g., ```json ... ```)
    text = text.replace("```json", "").replace("```", "").strip()

    try:
        return json.loads(text)
    except:
        return {"faithfulness": 0.0, "relevancy": 0.0}

# ==========================================
# 3. Helper Functions
# ==========================================
def compute_f1(prediction, truth):
    pred_tokens = prediction.lower().split()
    truth_tokens = truth.lower().split()
    common = collections.Counter(pred_tokens) & collections.Counter(truth_tokens)
    num_same = sum(common.values())
    if len(pred_tokens) == 0 or len(truth_tokens) == 0:
        return int(pred_tokens == truth_tokens)
    if num_same == 0: return 0.0
    precision = 1.0 * num_same / len(pred_tokens)
    recall = 1.0 * num_same / len(truth_tokens)
    return (2 * precision * recall) / (precision + recall)

def calculate_similarity(generated_ans, ground_truth_ans):
    emb1 = embed_model.encode([generated_ans])
    emb2 = embed_model.encode([ground_truth_ans])
    similarity = np.dot(emb1, emb2.T)[0][0]
    return float((similarity + 1) / 2)

def get_rag_response(question):
    return advanced_rag.query_with_analysis(question)

# ==========================================
# 4. Main Loop
# ==========================================
try:
    with open('ground_truth_100.json', 'r') as f:
        test_set = json.load(f)
except:
    test_set = []

results = []

if test_set:
    for i, item in enumerate(tqdm(test_set)):
        question = item['question']
        ground_truth_ans = item['ground_truth_answer']

        # --- REQUEST 1: GENERATE ANSWER ---
        rag_output = run_with_retry(get_rag_response, question)

        if rag_output:
            gen_ans = rag_output['answer']
            context_text = "\n".join([s['chunk'] for s in rag_output['sources']])

            # --- REQUEST 2: UNIFIED JUDGE (Faithfulness + Relevancy) ---
            # This replaces the previous 2 separate calls
            judge_scores = run_with_retry(combined_judge_call, question, gen_ans, context_text)

            if not judge_scores:
                judge_scores = {"faithfulness": 0.0, "relevancy": 0.0}

            # --- LOCAL METRICS (No API needed) ---
            sem_score = calculate_similarity(gen_ans, ground_truth_ans)
            f1_score = compute_f1(gen_ans, ground_truth_ans)
            citation_score = 1.0 if len(rag_output['citations']) > 0 else 0.0
            conf_score = rag_output['confidence']

            results.append({
                "id": item.get('id', i),
                "topic": item.get('topic', 'General'),
                "question": question,
                "ai_answer": gen_ans,
                "metrics": {
                    "Semantic Similarity": sem_score,
                    "F1 Score": f1_score,
                    "Faithfulness": judge_scores['faithfulness'],
                    "Answer Relevancy": judge_scores['relevancy'],
                    "Citation Presence": citation_score,
                    "Confidence": conf_score
                }
            })

            # Sleep is still good practice, but we can reduce it slightly
            time.sleep(1.5)
        else:
            print(f"‚ö†Ô∏è Skipping Q{i} due to generation error.")

    # ==========================================
    # 5. Final Report
    # ==========================================
    print("\n" + "="*60)
    print("üìä FINAL RAG PERFORMANCE REPORT")
    print("="*60)

    if results:
        metrics_list = results[0]['metrics'].keys()
        averages = {m: np.mean([r['metrics'][m] for r in results]) for m in metrics_list}

        hallucinated_count = len([r for r in results if r['metrics']['Faithfulness'] < 0.5])
        hallucination_rate = hallucinated_count / len(results)

        print(f"1Ô∏è‚É£  Generative Accuracy:")
        print(f"   ‚Ä¢ Semantic Similarity: {averages['Semantic Similarity']:.2%}")
        print(f"   ‚Ä¢ F1 Score:            {averages['F1 Score']:.2%}")
        print(f"   ‚Ä¢ Answer Relevancy:    {averages['Answer Relevancy']:.2%}")

        print(f"\n2Ô∏è‚É£  Trust & Grounding:")
        print(f"   ‚Ä¢ Faithfulness:        {averages['Faithfulness']:.2%}")
        print(f"   ‚Ä¢ Citation Rate:       {averages['Citation Presence']:.2%}")
        print(f"   ‚Ä¢ Hallucination Rate:  {hallucination_rate:.2%}")

        print(f"\n3Ô∏è‚É£  Retrieval Quality:")
        print(f"   ‚Ä¢ System Confidence:   {averages['Confidence']:.2%}")

        with open('final_evaluation_metrics.json', 'w') as f:
            json.dump(results, f, indent=2)
        print("\nüíæ Results saved to 'final_evaluation_metrics.json'")
    else:
        print("‚ùå No results collected.")
````
